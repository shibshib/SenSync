```python
import pandas as pd
from datetime import datetime as dt

## read crystal- and dco- timestamped data
columns = ['Timestamp', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'Gx', 'Gy', 'Gz', 'Ax', 'Ay', 'Az', 'Active'];
positive_emg_df = pd.read_csv('Data\EMG\Positive_dataset.csv', names=columns);
p_emg = positive_emg_df.as_matrix();

# Convert timestamps to epoch time
epoch = dt(1970, 1, 1)
for i in range(0, len(p_emg)):
    otime = p_emg[i][0];
    epoch_time = int((dt.strptime(otime, "%a %B %d %H:%M:%S %Y") - epoch).total_seconds())* 1000.0
    p_emg[i][0] = epoch_time;


## read crystal- and dco- timestamped data
columns = ['Timestamp', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'Gx', 'Gy', 'Gz', 'Ax', 'Ay', 'Az', 'Active'];
negative_emg_df = pd.read_csv('Data\EMG\\Negative_dataset.csv', names=columns);
n_emg = negative_emg_df.as_matrix();

# Convert timestamps to epoch time
epoch = dt(1970, 1, 1)
for i in range(0, len(n_emg)):
    otime = n_emg[i][0];
    epoch_time = int((dt.strptime(otime, "%a %B %d %H:%M:%S %Y") - epoch).total_seconds()) * 1000.0
    n_emg[i][0] = epoch_time;
```

```python
from scipy import signal
# Quick define a high pass filter
def butter_highpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def butter_highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = signal.filtfilt(b, a, data)
    return y

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = signal.filtfilt(b, a, data)
    return y
```
